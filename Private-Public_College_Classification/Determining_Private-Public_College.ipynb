# Goal

```python
# Import the libraries
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import svm 
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Import the dataset
df_college = pd.read_csv('College.csv')
df_college.head()
```

```python
# Check data types
df_college.info()
```

```python
# Explore the data
df_college.describe()
```

```python
# Check for missing numbers
df_college.isnull().sum()
```

```python
# No missing values
# Print the shape and the number of male/female
print('Shape of data',df_college.shape)
print('Total number of labels: {}'.format(df_college.shape[0]))
```

```python
df_college.Private.nunique()
```

```python
# Choose the features
X = df_college.drop(['Private'],axis=1)
print(df_college.shape)
print(X.shape)
```

```python
# Label encode the target, which is a categorical data
y = df_college['Private']
private_encoder = LabelEncoder()
y = private_encoder.fit_transform(y)
y
```

```python
# Train the data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
svc_model = SVC()
svc_model.fit(X_train,y_train)
y_pred = svc_model.predict(X_test)
```

```python
print('Accuracy score:\n', metrics.accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Let's standardize the data
# Scale the features
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
# Train the data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=42)
svc_model = SVC()
svc_model.fit(X_train,y_train)
y_pred = svc_model.predict(X_test)
print('Accuracy score:\n', metrics.accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
# See if hyperparameter optimization improves the result
param_grid = {'C':[0.1,1,10,100],
             'gamma': [1,0,1,0.01,0.001],
             'probability': [False,True],
             }

grid_search = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid_search.fit(X_train,y_train)
```

```python
print(grid_search.best_params_)
```

```python
print(grid_search.best_score_)
```

```python
grid_predictions = grid_search.predict(X_test)
```

```python
print('Accuracy score:\n', metrics.accuracy_score(y_test,grid_predictions))
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```

```python
## Perform nonlinear SVM - For this, we do hyperparameter optimization right away
# See if hyperparameter optimization improves the result
param_grid = {'gamma': [1,0,1,0.01,0.001],
              'nu': [0.2,0.3,0.5,0.8,1.0],
              'coef0':[0.0,0.2,0.5],
             'probability': [False,True],
              'tol': [0.001,0.0001],
              'decision_function_shape':['ovo', 'ovr'],
             }

grid_search = GridSearchCV(svm.NuSVC(kernel='poly'),param_grid,refit=True,verbose=2)
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
grid_predictions = grid_search.predict(X_test)
```

```python
print('Accuracy score:\n', metrics.accuracy_score(y_test,grid_predictions))
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
